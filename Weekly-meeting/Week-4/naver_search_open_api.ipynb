## 네이버 검색 api
 - https://developers.naver.com/docs/search/blog/

```python
# 네이버 검색 Open API 예제 - 블로그 검색
import os
import sys
import json
import urllib.request
```

```python
client_id = "YOUR ID"
client_secret = "YOUR SECRET"
```

```python
encText = urllib.parse.quote("강남역 맛집")
```

```python
encText
```

```python
num = 100
num_display = "&display={}".format(str(num))
```

```python
url = "https://openapi.naver.com/v1/search/blog.json?query=" + encText + num_display # json 결과

request = urllib.request.Request(url)
request.add_header("X-Naver-Client-Id",client_id)
request.add_header("X-Naver-Client-Secret",client_secret)
response = urllib.request.urlopen(request)
rescode = response.getcode()
```

```python
url
```

```python
rescode
```

```python
if(rescode==200):
    response_body = response.read()
    print(response_body.decode('utf-8'))
else:
    print("Error Code:" + rescode)
```

```python
type(response_body.decode('utf-8'))
```

```python
json_body = json.loads(response_body.decode('utf-8'))
```

```python
json_body
```

```python
import pandas as pd
import re
```

```python
json_body['items']
```

```python
title_list = []
link_list = []
name_list = []
date_list = []
desc_list = []

for line in json_body['items']:
    title_list.append(re.sub("(<b>|</b>)","", line['title']))
    link_list.append(line['link'].replace("\\", ""))
    name_list.append(line['bloggername'])
    date_list.append(line['postdate'])
    desc_list.append(re.sub("(<b>|</b>)","", line['description']))
```

```python
df = pd.DataFrame({
                   "post date": date_list,
                   "blogger name": name_list,
                   "title": title_list,
                   "contents": desc_list,
                   "link": link_list,
                  })
```

## Papago Open API
- https://developers.naver.com/docs/papago/papago-nmt-overview.md

```python
import os
import sys
import urllib.request
```

```python
client_id = "YOUR ID"
client_secret = "YOUR SECRET"
```

```python
encText = urllib.parse.quote("요즘 어린이들은 유튜브로 뽀로로를 즐겨봅니다")
```

```python
encText
```

```python
data = "source=ko&target=en&text=" + encText
url = "https://openapi.naver.com/v1/papago/n2mt"
```

```python
request = urllib.request.Request(url)
request.add_header("X-Naver-Client-Id",client_id)
request.add_header("X-Naver-Client-Secret",client_secret)
response = urllib.request.urlopen(request, data=data.encode("utf-8"))
rescode = response.getcode()
```

```python
rescode
```

```python
if(rescode==200):
    response_body = response.read()
    print(response_body.decode('utf-8'))
else:
    print("Error Code:" + rescode)
```
