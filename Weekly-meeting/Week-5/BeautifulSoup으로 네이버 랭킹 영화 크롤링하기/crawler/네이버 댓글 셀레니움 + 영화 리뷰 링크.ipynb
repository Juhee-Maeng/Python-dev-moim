# 네이버 댓글 크롤링

```python
import pandas as pd
from selenium import webdriver
#시간을 찍어보기 위해 time 사용
import time

url = 'https://news.naver.com/main/read.nhn?mode=LPOD&mid=sec&oid=009&aid=0004639995'

#웹 드라이버
driver = webdriver.Chrome('./chromedriver.exe')

#인터넷 상황이 안좋다면 15~30
driver.implicitly_wait(5)
driver.get(url)
```

```python
#아래 코드는 더보기라는 버튼을 누르기 위함
button = driver.find_element_by_css_selector('span.u_cbox_in_view_comment')
button.click()
while True:
    try:
        button = driver.find_element_by_css_selector('span.u_cbox_page_more')
        button.click()
        time.sleep(1)
    except:
        break
print('끝')
```

```python
#댓글 추출
contents = driver.find_elements_by_css_selector('span.u_cbox_contents')
for content in contents:
    print(content.text)
```

# 네이버 영화 리뷰 링크 크롤링

```python
# 웹페이지에 소스코드를 파싱하기
from bs4 import BeautifulSoup

driver.get("https://movie.naver.com/movie/sdb/rank/rmovie.nhn?sel=pnt&date=20200820&page=1")
html = driver.page_source
soup = BeautifulSoup(html, 'lxml')

# 파싱
title_list = soup.find_all('div', 'tit5')
```

```python
title_list
```

### 영화 제목과 해당 링크를 파싱

```python
# 제목 출력
movie_title = []
link_list = []

# 텍스트만 추출, 공백 제거
for title in title_list:
    movie_title.append(title.text.strip())
    result = title.find('a')
    result_data = 'https://movie.naver.com/' + result.get('href')
    link_list.append(result_data)
```

```python
link_list
```

```python
dic = { movie_title:link_list for movie_title, link_list in zip(movie_title, link_list) }
```
