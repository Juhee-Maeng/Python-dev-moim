```python
import argparse
```

```python
parser = argparse.ArgumentParser() #객체 생성
```

- help는 도움말을 뜻함 
- required = True는 인자값을 전달 안하고 실행시키면 자동으로 help문이 뜨게 함

```python
parser.add_argument('--src', '-s', help="ip src address", required=True)
args = parser.parse_args()
```

```python
print(args.src)
```
